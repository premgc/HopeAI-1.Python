```python
#pandas is a library and it contains the functions related to the dataframe
import pandas as pd
```

```python
#dataset is loaded and read into memory variable called dataset 
# read_csv is the function take file name as parameter finally 
# creates pandas DataFrame
dataset=pd.read_csv("Salary_Data.csv")
 
```

```python
#Display Pandas DataFrame loaded into the variable dataset in above cell[4]
dataset
```

```python

# using the index (row labels) of the DataFrame. slice and dice dataset
independent=dataset[["YearsExperience"]]#
independent 
```

```python
# using the  index (row labels) of the DataFrame. . slice and dice dataset
dependent=dataset[["Salary"]]
dependent
```

```python

# libraries are imported from class sklearn.
## specfic method is imported train_test_split
# dataset independent, dependent passed into 
## respective variables X_train, y_train loaded into the model
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(independent, dependent, test_size=0.30,random_state=0)
```

```python
# libraries are imported from class sklearn.
## LinearRegression function is imported 
## LinearRegression() is stored in varaiable regressor
## regresson.fit method invoked (X_train, y_regressor.coef_train) dataset is passed

from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(X_train,y_train)
```

```python
### The coefficients Attributes
weight=regressor.coef_
weight
```

```python
### The intercept_ Attributes
bais=regressor.intercept_
bais
```

```python
##prediction
y_pred=regressor.predict(X_test)
```

```python
##import r2_score from class sklearn function metrics
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
```

```python
##varibale data is displayed from above step [38]
r_score
```

```python
## import pickle to save the model filename 'finalized_model_linear.sav' 
## varabile value is set.
import pickle
filename="finalized_model_linear.sav"
```

```python
## to save the model to the given filename write binary
pickle.dump(regressor,open(filename,'wb'))
```

```python

## load the model read binary filename='finalized_model_linear.sav' 
loaded_model=pickle.load(open("finalized_model_linear.sav",'rb'))
result=loaded_model.predict([[13]])
```
